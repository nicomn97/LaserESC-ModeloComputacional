```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
```

```python
df = pd.read_csv('Registros/datosMatlab.csv', delimiter=',', decimal='.')
data = df.loc[(df['t']>=20) & (df['t']<=40)]
```

```python
dfPicoError = pd.read_csv('Registros/senalErrorPico.csv', delimiter=',', decimal='.', header=None,  skiprows=18)
dfError = pd.read_csv('Registros/senalError.csv', delimiter=',', decimal='.', header=None,  skiprows=18)

t=dfError[3]
Vpd=dfPicoError[4]
Verr=dfError[4]


t0=500
tf=-50
```

```python
matplotlib.rc('xtick', labelsize=24) 
matplotlib.rc('ytick', labelsize=24) 
```

```python
fig, ax = plt.subplots(1,2, sharex=True, sharey=True, figsize=(12,5))

plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

ax[0].plot(data['t']-20, data['Vpd'], c='r', label='Control Activo', linewidth=2)
ax[0].set_ylabel(r'$V_{PD}$ [V]', size=24)
ax[0].set_ylim([0.94, 1.175])
ax[0].grid()
ax[0].set_xlabel(r'tiempo [s]', size=24)
ax[0].set_title(r'Simulaci√≥n', size=24)


ax[1].plot(t[t0:tf]-t[t0],Vpd[t0:tf], c='b', label='Control Activo', linewidth=3)
ax[1].set_ylim([0.94, 1.175])
ax[1].set_xlabel(r'tiempo [s]', size=24)
ax[1].grid()
ax[1].set_title(r'Experimento', size=24)
plt.tight_layout()
plt.savefig('comparacionSenalesEspec.png')
```
