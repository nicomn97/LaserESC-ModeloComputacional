```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
```

```python
c=299792458
```

```python
simulacionEstable = pd.read_csv('Registros/simulacionEstable.csv', delimiter=',', decimal='.')
simulacionInestable = pd.read_csv('Registros/simulacionInestable.csv', delimiter=',', decimal='.')

expEstable = pd.read_csv('Registros/expEstable.txt', delimiter='\t', skiprows=5, decimal=',', names=['tiempo','Serror','Vpd','wl'])
expInestable = pd.read_csv('Registros/expInestable.lta', delimiter='\t', skiprows=54, decimal=',', names=['time','wl','power'])
```

```python
matplotlib.rc('xtick', labelsize=24) 
matplotlib.rc('ytick', labelsize=24) 
```

```python
fig, ax = plt.subplots(1,2, sharex=True, sharey=True, figsize=(12,5))

plt.xticks(fontsize=24)
plt.yticks(fontsize=24)

ax[0].plot(simulacionEstable['t']/3, simulacionEstable['nu'], c='r', label='Control Activo', linewidth=4)
ax[0].set_ylabel(r'$\nu_{L} - \nu_{5}$ [MHz]', size=24)
ax[0].set_ylim([-80, 80])
ax[0].grid()
ax[0].set_xlabel(r'tiempo [h]', size=24)
ax[0].set_title(r'Simulaci√≥n', size=24)


ax[1].plot(expEstable['tiempo']/(1000*60*60), (1000*c/(expEstable['wl']))-(1000*c/(expEstable['wl'].mean())), c='b', label='Control Activo')
ax[1].set_ylim([-80, 80])
ax[1].set_xlabel(r'tiempo [h]', size=24)
ax[1].grid()
ax[1].set_title(r'Experimento', size=24)
plt.tight_layout()

ax[0].set_xlim(-0.1,2.1)
ax[1].set_xlim(-0.1,2.1)


plt.savefig('comparacionEstable.png')
```
